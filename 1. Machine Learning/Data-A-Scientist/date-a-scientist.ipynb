```python
for value in education_order:
    if value not in list(education_mappings.keys()):
        print(value)
```

```python
list_of_ethnicities = []
value_list = profiles['ethnicity'].value_counts().index.tolist()
for value in profiles['ethnicity'].value_counts().index.tolist():
    if ", " in value:
        value_list = value.split(", ")
        for value2 in value_list:
            if value2 not in list_of_ethnicities:
                list_of_ethnicities.append(value2)
    else:
        if value not in list_of_ethnicities:
            list_of_ethnicities.append(value)
list_of_ethnicities
```

```python
for ethnicity in list_of_ethnicities:
    profiles[ethnicity] = 0
    profiles[ethnicity] = profiles['ethnicity'].apply(lambda row: 1 if ethnicity in row else 0)
```

## LogisticRegression

```python
features = profiles[['drinks', 'drugs', 'education', 'cats', 'dogs']]
labels = profiles['sex']
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data, train_labels, test_labels = train_test_split(features, labels, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(train_data, train_labels)
predicted_labels = model.predict(test_data)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(test_labels, predicted_labels)
print(f"Logistic Model Accuracy: {accuracy}")
```
