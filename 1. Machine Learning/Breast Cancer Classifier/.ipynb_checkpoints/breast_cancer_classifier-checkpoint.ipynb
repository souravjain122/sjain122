# Breast Cancer Classifier

## Introduction
After skin cancer, breast cancer is the most common cancer (basically 2nd most common cancer) diagnosed in women in the United States. Breast cancer can occur in both men and women, but it's far more common in women. Just like skin cancer its treatment involves chemotherapy, radiation, hormone therapy and surgery.
<br>
<br>
In this project, we will be using several Python libraries to make a K-Nearest Neighbor classifier that is trained to predict whether a patient has breast cancer.
<br>
And then we will also use Logistic Regression.

## Import Required Libraries

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

## Load Data

```python
(breast_cancer_data, breast_cancer_target) = load_breast_cancer(return_X_y = True, as_frame = True)
breast_cancer_data
```

```python
breast_cancer_target
```

## Investigate Data

```python
breast_cancer_data.isnull().sum()
```

```python
breast_cancer_target.isnull().sum()
```

Great! We don't have to worry about null values now in data and target.

## Check for Multicolinearity

```python
plt.figure(figsize = (18, 10))
corr_grid = breast_cancer_data.corr()
sns.heatmap(corr_grid,
            xticklabels = corr_grid.columns,
            yticklabels = corr_grid.columns,
            vmin = -1, center = 0, vmax = 1,
            annot = True, cmap = "PuOr")
plt.show()
plt.clf()
```

Based on above multicolinearity check we will be droping
- `worst concave points`
- `worst area`
- `worst perimeter`
- `worst texture`
- `worst radius`
- `area error`
- `perimeter error`
- `mean concavity`
- `mean area`
- `mean perimeter`

```python
breast_cancer_data = breast_cancer_data.drop(columns = ['worst concave points', 'worst area', 'worst perimeter', 'worst texture', 'worst radius', 'area error', 'perimeter error', 'mean concavity', 'mean area', 'mean perimeter'])
breast_cancer_data
```

```python
plt.figure(figsize = (18, 10))
corr_grid = breast_cancer_data.corr()
sns.heatmap(corr_grid,
            xticklabels = corr_grid.columns,
            yticklabels = corr_grid.columns,
            vmin = -1, center = 0, vmax = 1,
            annot = True, cmap = "PuOr")
plt.show()
plt.clf()
```

## Splitting data in training and validation data

```python
training_data, validation_data, training_labels, validation_labels = train_test_split(breast_cancer_data, breast_cancer_target, test_size = 0.2, random_state = 100)
print(len(training_data))
print(len(training_labels))
```

## Predicting outcomes for k in range of 1 to 100 (inclusive)

```python
max_score = 0
max_k = 1
accuracies = []
for k in range(1,101):
    classifier = KNeighborsClassifier(k)
    classifier.fit(training_data, training_labels)
    accuracy = classifier.score(validation_data, validation_labels)
    accuracies.append(accuracy)
    print(f"for k = {k}, the classifier score is {round(accuracy, 5)}")
    if max_score < accuracy:
        max_score = accuracy
        max_k = k
```

```python
print(f"K for max score and max scores are {max_k}, {max_score} respectively")
```

## Visual representation

```python
k_list = list(range(1,101))
plt.plot(k_list, accuracies)
plt.xlabel("k")
plt.ylabel("Validation Accuracy")
plt.title("Breast Cancer Classifier Accuracy")
plt.show()
```

## Prediction for k = 26

```python
k = 26
classifier = KNeighborsClassifier(k)
classifier.fit(training_data, training_labels)
prediction_labels = classifier.predict(validation_data)
prediction_labels
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_fscore_support
```

```python
accuracy = accuracy_score(validation_labels, prediction_labels)
precision, recall, f1_score, support = precision_recall_fscore_support(validation_labels, prediction_labels,
                                                                       pos_label = 1, average = "binary")
# pos_label = "xyz" meaning "xyz" is True value
# average determine the type of averaging to be performed. In our case it is binary
print(f"Logistic Model Accuracy: {accuracy}")
print(f"Logistic Model Precision: {precision}")
print(f"Logistic Model Recall: {recall}")
print(f"Logistic Model F1-score: {f1_score}")
print(f"Logistic Model Support: {support}")
```

# Logistic Regression

## Import Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

## Create and Fit the Model

```python
model = LogisticRegression()
model.fit(training_data, training_labels)
```

# Model Coefficients and Intercept

```python
training_data.columns.tolist()
```

```python
coefficients = model.coef_

columns = training_data.columns.tolist()
print("Model Coefficients are:")
for i in range(len(columns)):
    print(f"{columns[i]}: {coefficients[0][i]}")
```

```python
intercept = model.intercept_
print(f"Model Intercept is: {intercept}")
```

```python
prediction_labels = model.predict(validation_data)
prediction_labels
```

```python
accuracy = accuracy_score(validation_labels, prediction_labels)
precision, recall, f1_score, support = precision_recall_fscore_support(validation_labels, prediction_labels,
                                                                       pos_label = 1, average = "binary")
# pos_label = "xyz" meaning "xyz" is True value
# average determine the type of averaging to be performed. In our case it is binary
print(f"Logistic Model Accuracy: {accuracy}")
print(f"Logistic Model Precision: {precision}")
print(f"Logistic Model Recall: {recall}")
print(f"Logistic Model F1-score: {f1_score}")
print(f"Logistic Model Support: {support}")
```
