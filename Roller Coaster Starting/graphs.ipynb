### Data Visualizations

__Data Visualization__ is an important component for __Exploratory Data Analysis__ because it allows the Analyst to:
1. Look at the distribution of Data
2. Understand the relationship between variables
3. Find the pattern (if present) in the Data

and a many of other questions can be answered through __Data Visualization__

This time our approach for Data Visualizations would be different from other projects. This time we will create all the functions necessary for ploting and make plots later. A function for:
1. Rank vs Year of Rank
2. Top N Rankings vs Year of Rank
3. Histograms
4. Number of Inversions for Roller Coasters in a specified Park
5. Piechart for Operating and Definitely Closed roller coasters
6. Scatter plot for any two given columns

```python
# Rank vs Year of Rank
def coaster_rank(df, *names):
    number_of_names = len(names)
    title = ""
    i = 0
    plt.figure(figsize = (15, 4))
    for name in names:
        i += 1
        x = df['Year of Rank'][df['Name'] == name]
        y = df['Rank'][df['Name'] == name]
        plt.plot(x, y)
        if i < number_of_names:
            title = title + str(name) + " vs "
        elif i == number_of_names:
            title = title + str(name) + " Rankings"
    plt.xlabel("Year")
    plt.ylabel("Ranking")
    plt.title(f"{title}")
    plt.show()
    plt.clf()

# Top N Rankings vs Year of Rank
def top_n_coaster_rank(n, df):
    top_n_rankings = df[df['Rank'] <= n]
    plt.figure(figsize = (15, 4))
    for name in set(top_n_rankings['Name']):
        rankings = top_n_rankings[top_n_rankings['Name'] == name]
        x = rankings['Year of Rank']
        y = rankings['Rank']
        plt.plot(x, y, label = name)
    plt.xlabel("Year")
    plt.ylabel("Rankings")
    plt.title(f"Top {n} Rankings")
    plt.legend(bbox_to_anchor = (1, 1))
    plt.show()
    plt.clf()

# Histograms
def plot_hist(df, name):
    new_df = df
    new_df = new_df.dropna(subset = [name])
    plt.figure(figsize = (15, 4))
    plt.hist(new_df[name])
    plt.xlabel(f"{name}")
    plt.ylabel("Number of Roller Coasters")
    plt.show()
    plt.clf()

# Number of Inversions for Roller Coasters in a specified Park
def plot_bar(coaster_df, park_name):
    park_coasters = coaster_df[coaster_df['park'] == park_name]
    park_coasters = park_coasters.sort_values('num_inversions', ascending = False)
    coaster_names = park_coasters['name']
    number_inversions = park_coasters['num_inversions']
    coaster_names_index = [i for i, name in enumerate(coaster_names)]
    plt.figure(figsize = (15, 4))
    plt.bar(coaster_names_index, number_inversions)
    plt.xticks(coaster_names_index, coaster_names, rotation = "vertical")
    plt.xlabel("Coaster Name")
    plt.ylabel("Number of Inversions")
    plt.show()
    plt.clf()

# Piechart for Operating and Definitely Closed roller coasters
def plot_pie(df):
    value_counts = df['status'].value_counts()
    data = value_counts.tolist()[:2]
    labels = value_counts.index.tolist()[:2]
    plt.figure(figsize = (15, 4))
    plt.pie(data, labels = labels)
    plt.show()
    plt.clf()

# Scatter plot for any two given columns
def plot_scatter(df, name1, name2):
    x = df[name1]
    y = df[name2]
    plt.figure(figsize = (15, 4))
    plt.scatter(x, y)
    plt.show()
    plt.clf()
```

# Start Plottings

```python
# Create a plot of El Toro ranking over time
coaster_rank(wood, "El Toro")
# Create a plot of El Toro and Boulder dash hurricanes
coaster_rank(wood, "El Toro", "Boulder Dash")
```

```python
# Create a plot of top n rankings over time
top_n_coaster_rank(5, wood)
```
